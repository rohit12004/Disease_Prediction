```python
X = df.drop('prognosis', axis=1)
y = df['prognosis']
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Convert disease names to numerical labels
y_encoded = label_encoder.fit_transform(y)

# Check the mapping of diseases to numbers
disease_mapping = dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))
print("Disease Mapping:\n", disease_mapping)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model with the encoded disease labels
model.fit(X, y_encoded)

print("Model training completed with encoded labels!")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

model2 = RandomForestClassifier(n_estimators=100, random_state=42)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X,y_encoded, test_size=0.2, random_state=42)

# Train the model on the training data
model2.fit(X_train, y_train)
```

```python
y_pred = model2.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(accuracy*100)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  # Create a scaler object
X_train_scaled = scaler.fit_transform(X_train)  # Scale X_train
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(RandomForestClassifier(), X_train_scaled, y_train, cv=5, scoring='accuracy')
cv_scores.mean()*100
```

```python
input_vector = [0] * len(symptoms_list)
```

```python
selected_symptoms = ['continuous_sneezing', 'shivering', 'chills', '0', '0']
```

```python
for symptom in selected_symptoms:
    if symptom != "0" and symptom in symptoms_list:  # Ignore "0" and only process valid symptoms
        index = symptoms_list.get_loc(symptom)  # Use get_loc to find index in Index object
        input_vector[index] = 1
```
